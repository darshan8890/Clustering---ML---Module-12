# **Clustering Assignment Questions**

# **Theoretical Questions**

Q1. What is unsupervised learning in the context of machine learning ?
- Unsupervised learning is a type of machine learning where the model learns from data without any labels. It tries to find patterns, groups, or structures in the data on its own. A common example is clustering, where similar items are grouped together.

Q2. How does K-Means clustering algorithm work ?
- K-Means clustering is an unsupervised learning algorithm used to group similar data points into clusters. It starts by choosing a number of clusters (K) and randomly selecting K points as the initial cluster centers, called centroids.
- Then, it assigns each data point to the nearest centroid, forming groups. After that, it updates the centroids by calculating the average position of all the points in each group. This process of assigning points and updating centroids continues until the centroids no longer move much, meaning the clusters have become stable. The result is that the data is divided into K groups based on similarity.

Q3. Explain the concept of a dendrogram in hierarchical clustering.
- A dendrogram is a tree-like diagram used to show the arrangement of clusters formed by hierarchical clustering. It visually represents how data points are grouped together step by step. At the bottom of the dendrogram, each individual data point starts as its own cluster. As you move up the diagram, similar clusters are gradually merged together.
- The height at which two clusters are joined shows how similar or different they are - lower connections mean more similar, and higher connections mean less similar. By cutting the dendrogram at a certain height, you can choose the number of clusters you want from the data.

Q4. What is the main difference between K-Means and Hierarchical Clustering?
- K-Means clustering = K-Means clustering is a partition-based method that requires you to specify the number of clusters (K) in advance. It works by randomly selecting K centroids and then iteratively assigning data points to the nearest centroid and updating the centroids based on the mean of the points in each cluster. It is fast and efficient for large datasets but can be sensitive to the initial placement of centroids and may not perform well with complex cluster shapes.
- hierarchical Clustering = Hierarchical clustering, on the other hand, does not require you to choose the number of clusters at the start. It builds a hierarchy of clusters either by merging smaller clusters into bigger ones (agglomerative) or by splitting larger clusters into smaller ones (divisive). The results are shown in a dendrogram, which helps visualize how clusters are formed step by step. This method is more flexible and informative but can be slower and more memory-intensive, especially with large datasets.

Q5.  What are the advantages of DBSCAN over K-Means ?
- DBSCAN has some key advantages over K-Means. First, it does not require you to choose the number of clusters in advance. Second, it can find clusters of any shape, while K-Means works best with round or evenly sized clusters. Third, DBSCAN is better at handling noise and outliers, as it can mark them separately instead of forcing them into a cluster.

Q6. When would you use Silhouette Score in clustering ?
- You would use the Silhouette Score to measure how well your data has been clustered. It tells you how similar each point is to its own cluster compared to other clusters. A higher score (close to 1) means the point is well matched to its own cluster and poorly matched to others. You can use it to choose the best number of clusters or to check the quality of your clustering results.

Q7. What are the limitations of Hierarchical Clustering?
- Hierarchical clustering has some limitations. It is slow with large datasets and uses a lot of memory. Once clusters are joined, it can’t undo mistakes, which can lead to poor results. It is also sensitive to noise and outliers, and the final clusters can change depending on the method used to measure distances between clusters.

Q8. Why is feature scaling important in clustering algorithms like K-Means ?
- Feature scaling is important in clustering because it ensures all features are treated equally. Without scaling, features with larger values can unfairly affect the clustering results, leading to incorrect groups. Scaling makes sure every feature contributes fairly to the clustering process.

Q9. How does DBSCAN identify noise points ?
- DBSCAN identifies noise points as those that don’t have enough nearby points to form a cluster. These points are neither core points nor border points and are marked as noise.

Q10. Define inertia in the context of K-Means.
- Inertia in K-Means measures how spread out the points are within each cluster. It calculates the sum of squared distances between points and their cluster center. Lower inertia means better, tighter clusters.

Q11. What is the elbow method in K-Means clustering?
- The elbow method helps find the best number of clusters for K-Means. You plot how the error (distance between data points and cluster centers) changes as you increase the number of clusters. The "elbow" is the point where the error starts decreasing more slowly, showing the ideal number of clusters to use.

Q12. Describe the concept of "density" in DBSCAN.
- In DBSCAN, "density" means how close the points are to each other. If enough points are within a certain distance (Epsilon) and meet a minimum number (MinPts), they form a cluster. Points that don’t meet these conditions are considered noise or outliers.

Q13. Can hierarchical clustering be used on categorical data ?
- hierarchical clustering can handle categorical data by using appropriate distance metrics, like the Hamming distance or Gower's similarity coefficient, which are designed for categorical variables. However, clustering results depend on how well these metrics capture the relationships between data points.

Q14. What does a negative Silhouette Score indicate?
- A negative Silhouette Score indicates that a data point is likely closer to points in another cluster than to points in its own cluster. This suggests that the point may be misclassified or that the clustering structure doesn't fit the data well. It can also highlight overlapping or poorly separated clusters.

Q15. Explain the term "linkage criteria" in hierarchical clustering.
- Linkage criteria in hierarchical clustering define how the distance between clusters is measured to decide which clusters should be merged. Single linkage calculates the distance using the nearest points between clusters, while complete linkage uses the farthest points.
- Average linkage focuses on the average distance between all points in the clusters, and Ward's method minimizes the internal differences within clusters by reducing overall variance. Each method leads to distinct clustering results based on how it interprets the data's relationships.


Q16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
- K-Means clustering can perform poorly on data with clusters of varying sizes or densities because it assumes clusters are similar in size, shape, and density. Larger clusters may overshadow smaller ones, causing smaller clusters to merge incorrectly.
- Dense clusters might be divided into multiple parts, while overlapping clusters can lead to misclassification. These limitations arise because K-Means relies heavily on distance-based calculations, which may not accurately represent the diverse structures of such data.

Q17.  What are the core parameters in DBSCAN, and how do they influence clustering ?
- DBSCAN has two key parameters: epsilon (ε) and MinPts.
- Epsilon = Defines the maximum distance within which points are considered neighbors. Smaller values create tighter clusters, while larger values may merge clusters or include noise.
- MinPts: Specifies the minimum number of points required to form a dense region. Higher values result in fewer, larger clusters, while lower values may create more clusters or include noise.


Q18. How does K-Means++ improve upon standard K-Means initialization?
- K-Means++ improves standard K-Means initialization by selecting initial cluster centers more strategically. Instead of choosing centers randomly, it picks the first center randomly and then selects subsequent centers based on their distance from already chosen centers.
- This ensures that the initial centers are well-spread out, reducing the chances of poor clustering and speeding up convergence. This method often leads to better clustering results compared to random initialization.


Q19. What is agglomerative clustering?
- Agglomerative clustering is a type of hierarchical clustering that starts with each data point as its own cluster. Gradually, it merges the closest clusters based on a chosen distance metric, continuing until all points are grouped into a single cluster or a stopping criterion is met. This bottom-up approach helps identify patterns and relationships in the data.

Q20. What makes Silhouette Score a better metric than just inertia for model evaluation?
- Silhouette Score is a better metric than inertia because it evaluates both cohesion (how close points are within a cluster) and separation (how far clusters are from each other).
- In contrast, inertia only measures compactness within clusters and decreases as the number of clusters increases, which can lead to overfitting. Silhouette Score provides a more balanced view, helping to assess the quality of clustering more effectively.

```python
# Q21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering with 4 Centers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
# Q29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.


from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=1000, centers=4, cluster_std=1.0, random_state=42)

silhouette_scores = []
k_values = range(2, 6)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

plt.figure(figsize=(8, 6))
plt.bar(k_values, silhouette_scores, color='skyblue')
plt.xticks(k_values)
plt.title("Silhouette Scores for Different Values of k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
# Q30. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.


from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data

linked = linkage(X, method='average')

plt.figure(figsize=(10, 7))
dendrogram(linked, labels=iris.target, leaf_rotation=90, leaf_font_size=10)
plt.title("Dendrogram using Average Linkage (Iris Dataset)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()
```

```python
# Q31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.


import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

X, y = make_blobs(n_samples=500, centers=3, cluster_std=1.5, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.predict(X)
centroids = kmeans.cluster_centers_

def plot_decision_boundaries(X, labels, centroids, model):

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                         np.arange(y_min, y_max, 0.05))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.4, cmap=ListedColormap(['yellow', 'skyblue', 'lightgreen']))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30, edgecolor='k')
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
    plt.title("K-Means Decision Boundaries")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()
    plt.show()

plot_decision_boundaries(X, labels, centroids, kmeans)
```

```python
# Q34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.


from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = load_breast_cancer()
X = data.data

inertia_values = []
k_values = range(2, 7)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(k_values, inertia_values, marker='o', linestyle='-', color='blue')
plt.title("Inertia Values for K-Means (Breast Cancer Dataset)")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

```python
# Q44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.


from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
import matplotlib.pyplot as plt
import numpy as np

iris = load_iris()
X = iris.data

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

silhouette_vals = silhouette_samples(X, labels)

y_ticks = np.arange(len(silhouette_vals))
sorted_silhouette_vals = np.sort(silhouette_vals)

plt.figure(figsize=(8, 6))
plt.barh(y_ticks, sorted_silhouette_vals, color='skyblue')
plt.title("Silhouette Coefficient for Each Sample (Iris Dataset)")
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Sample Index")
plt.grid(True)
plt.show()
```

```python
# Q47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.


import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.5, random_state=42)

dbscan = DBSCAN(eps=0.75, min_samples=5)
labels = dbscan.fit_predict(X)

unique_labels = np.unique(labels)
n_clusters = len(unique_labels[unique_labels >= 0])

n_noise_points = np.sum(labels == -1)

print(f"Number of clusters identified: {n_clusters}")
print(f"Number of noise points: {n_noise_points}")
```
